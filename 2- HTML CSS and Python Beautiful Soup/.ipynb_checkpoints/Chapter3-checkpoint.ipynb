```python
import requests
from bs4 import BeautifulSoup
url = 'https://en.wikipedia.org/w/index.php' + \
      '?title=List_of_Game_of_Thrones_episodes&oldid=802553687'
r = requests.get(url)
html_contents = r.text
html_soup = BeautifulSoup(html_contents, 'html.parser')
# Find the first h1 tag
first_h1 = html_soup.find('h1')
print(first_h1.name)     # h1
print(first_h1.contents) # ['List of ', [...], ' episodes']
print(str(first_h1))
# Prints out: <h1 class="firstHeading" id="firstHeading" lang="en">List of
#             <i>Game of Thrones</i> episodes</h1>
print(first_h1.text)       # List of Game of Thrones episodes
print(first_h1.get_text()) # Does the same
print(first_h1.attrs)
# Prints out: {'id': 'firstHeading', 'class': ['firstHeading'], 'lang': 'en'}
print(first_h1.attrs['id']) # firstHeading
print(first_h1['id'])       # Does the same
print(first_h1.get('id'))   # Does the same
print('------------ CITATIONS ------------')
# Find the first five cite elements with a citation class
cites = html_soup.find_all('cite', class_='citation')
print("this is  the lenght EEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEEE",type(cites), len(cites))
for citation in cites:
    print(citation.get_text())
    # Inside of this cite element, find the first a tag
    try:
        link = citation.find('a')
        print(' this is the type of thelink',type(link))
        # ... and show its URL
        #if there is no link in the <a> tag then it will ruturn None type , we can cater for this
        #try except block
        print(link.get('href'))
        print()
    except:
        print('no link found')
        continue
```

### Fetching table data from Wikipedia 

```python
import requests
from bs4 import BeautifulSoup
url = 'https://en.wikipedia.org/w/index.php' + \
      '?title=List_of_Game_of_Thrones_episodes&oldid=802553687'
r = requests.get(url)
html_contents = r.text
html_soup = BeautifulSoup(html_contents, 'html.parser')
# We'll use a list to store our episode list
episodes = []
# here we are using class_ variable because python has built-in keword for "class" which will 
#  be ambigous with HTML class keyword
ep_tables = html_soup.find_all('table', class_='wikiepisodetable')
print(" TableeeeeeeeeeeeeeeeeeeVVVVVVVVVeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeee", len(ep_tables))
```

```python
for table in ep_tables:
    headers = []
    rows = table.find_all('tr')
     # Start by fetching the header cells from the first row to determine
     # the field names
#     for header in table.find('tr'):
#         headers.append(header.text)
#     print("headers list", headers[0][1])
    #Then go through all the rows except the first one
    for row in table.find_all('tr')[1:]:
        values = []
        # And get the column cells, the first one being inside a th-tag
        for col in row.find_all(['th','td']):
            values.append(col.text)
        if values:
            episode_dict = {headers[i]: values[i] for i in range(len(values))}
            episodes.append(episode_dict)
# Show the results
for episode in episodes:
    print(episode)
```
